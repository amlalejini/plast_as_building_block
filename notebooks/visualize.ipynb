```python
# imports
%matplotlib inline
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# A few relevant settings/paths
exp_base_dir = "/Users/amlalejini/DataPlayground/plast_as_building_block"
processed_dir = os.path.join(exp_base_dir, "processed")
data_fpath = os.path.join(exp_base_dir, "iter2_fdom_200k.data")
```

```python
# Visualize normalized phenotype scores
fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
sns.boxplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q1"], order = ["Q1T1", "Q1T2", "Q1T3"])
sns.swarmplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q1"], order = ["Q1T1", "Q1T2", "Q1T3"])

fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
sns.boxplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q2"][df["seed_norm_phenotype_score"] == 100.0], order = ["Q2T1", "Q2T2", "Q2T3"])
sns.swarmplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q2"][df["seed_norm_phenotype_score"] == 100.0], order = ["Q2T1", "Q2T2", "Q2T3"])

fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
sns.boxplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q3"][df["seed_norm_phenotype_score"] == 100.0], order = ["Q3T1", "Q3T2", "Q3T3"])
sns.swarmplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q3"][df["seed_norm_phenotype_score"] == 100.0], order = ["Q3T1", "Q3T2", "Q3T3"])

fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
sns.boxplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q4"][df["seed_norm_phenotype_score"] == 100.0], order = ["Q4T1", "Q4T2", "Q4T3"])
sns.swarmplot(x = "treatment", y = "fdom_phenotype_score", data = df[df["question"] == "Q4"][df["seed_norm_phenotype_score"] == 100.0], order = ["Q4T1", "Q4T2", "Q4T3"])
```
