## Final Dominant Lineage Scores Over Time

```python
ts_fpath = os.path.join(exp_base_dir, "lineage_score_ts.csv")
```

```python
# Load data
ts_df = pd.read_csv(ts_fpath)
#q1ts_df = ts_df[ts_df["question"] == "Q1"]
q2ts_df = ts_df[ts_df["question"] == "Q2"]
q3ts_df = ts_df[ts_df["question"] == "Q3"]
q4ts_df = ts_df[ts_df["question"] == "Q4"]
```

```python
# Plot score over time
fig, ax = plt.subplots()
fig.set_size_inches(30, 15)
sns.tsplot(data=q4ts_df, time="update", unit="rep",
           condition="treatment", value="score")
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(30, 15)
sns.tsplot(data=q3ts_df, time="update", unit="rep",
           condition="treatment", value="score")
```

```python
cp = ['#1b9e77','#d95f02','#7570b3']
pal = sns.color_palette(cp)

fig, ax = plt.subplots()
fig.set_size_inches(30, 15)
sns.tsplot(data=q2ts_df, time="update", unit="rep",
           condition="treatment", value="score", color = {"Q2T1": pal[0], "Q2T2": pal[1], "Q2T3": pal[2]})
```
