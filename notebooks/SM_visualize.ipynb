# The Effects of Slip Mutations on the Evolution of Complexity


```python
# imports
%matplotlib inline
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# A few relevant settings/paths
exp_base_dir = "/Users/amlalejini/DataPlayground/slip_muts/iter_2"
data_fpath = os.path.join(exp_base_dir, "processed", "slipmuts_q1q2_iter2_fdom_200k.data")
```

```python
# Load data
df = pd.read_csv(data_fpath)
df
```

```python
df_mg100 = df[df["treatment"].str.contains("MinGen_100") == True]
df_mg0 = df[df["treatment"].str.contains("MinGen_0") == True]
```

```python
label_map = {"Slip_NONE": "Baseline", "HIGHMUT": "High Mutation Rate", "Slip_SCATTER": "Slip: Scatter", "Slip_NOP": "Slip: nop-X", "Slip_RAND": "Slip: Random", "Slip_SCRAM": "Slip: Scrambled", "Slip_DUP": "Slip: Duplicate"}

q1_mg100_order = ["Q1__Slip_NONE__MinGen_100", "Q1__HIGHMUT__MinGen_100","Q1__Slip_SCATTER__MinGen_100",
                   "Q1__Slip_NOP__MinGen_100", "Q1__Slip_RAND__MinGen_100", "Q1__Slip_SCRAM__MinGen_100",
                   "Q1__Slip_DUP__MinGen_100"]

q2_mg100_order = ["Q2__Slip_NONE__MinGen_100", "Q2__HIGHMUT__MinGen_100","Q2__Slip_SCATTER__MinGen_100",
                   "Q2__Slip_NOP__MinGen_100", "Q2__Slip_RAND__MinGen_100", "Q2__Slip_SCRAM__MinGen_100",
                   "Q2__Slip_DUP__MinGen_100"]


q1_mg0_order = ["Q1__Slip_NONE__MinGen_0", "Q1__HIGHMUT__MinGen_0","Q1__Slip_SCATTER__MinGen_0",
                   "Q1__Slip_NOP__MinGen_0", "Q1__Slip_RAND__MinGen_0", "Q1__Slip_SCRAM__MinGen_0",
                   "Q1__Slip_DUP__MinGen_0"]

q2_mg0_order = ["Q2__Slip_NONE__MinGen_0", "Q2__HIGHMUT__MinGen_0","Q2__Slip_SCATTER__MinGen_0",
                   "Q2__Slip_NOP__MinGen_0", "Q2__Slip_RAND__MinGen_0", "Q2__Slip_SCRAM__MinGen_0",
                   "Q2__Slip_DUP__MinGen_0"]

q1_mg100_labels = [label_map[t.split("__")[1]] for t in q1_mg100_order]
q2_mg100_labels = [label_map[t.split("__")[1]] for t in q2_mg100_order]
q1_mg0_labels = [label_map[t.split("__")[1]] for t in q1_mg0_order]
q2_mg0_labels = [label_map[t.split("__")[1]] for t in q2_mg0_order]
```

## Final Dominant Lineage Scores Over Time

```python
q1q2ts_fpath = os.path.join(exp_base_dir, "processed", "SM_q1q2_lineage_score_ts.csv")
```

```python
# # Load data
ts_df = pd.read_csv(q1q2ts_fpath)
q1ts_df = ts_df[ts_df["question"] == "Q1"]
q2ts_df = ts_df[ts_df["question"] == "Q2"]

q1mg100ts_df = q1ts_df[q1ts_df["treatment"].str.contains("MinGen_100") == True]
q2mg100ts_df = q2ts_df[q2ts_df["treatment"].str.contains("MinGen_100") == True]
q1mg0ts_df = q1ts_df[q1ts_df["treatment"].str.contains("MinGen_0") == True]
q2mg0ts_df = q2ts_df[q2ts_df["treatment"].str.contains("MinGen_0") == True]
```

```python
colors = ['#a6cee3','#1f78b4','#b2df8a','#33a02c','#fb9a99','#e31a1c','#fdbf6f']
q1_mg100_cmap = {q1_mg100_order[con]:colors[con] for con in range(0, len(q1_mg100_order))}
q2_mg100_cmap = {q2_mg100_order[con]:colors[con] for con in range(0, len(q2_mg100_order))} 
q1_mg0_cmap = {q1_mg0_order[con]:colors[con] for con in range(0, len(q1_mg0_order))} 
q2_mg0_cmap = {q2_mg0_order[con]:colors[con] for con in range(0, len(q2_mg0_order))} 
```

### Static Logic 9 Min Genome 100

```python
# # Plot score over time
fig, ax = plt.subplots()
fig.set_size_inches(30, 15)
sns.tsplot(data=q1mg100ts_df, time="update", unit="rep",
           condition="treatment", value="score", color=q1_mg100_cmap)
```

### Static Logic 9 Min Genome 0

```python
# # Plot score over time
fig, ax = plt.subplots()
fig.set_size_inches(30, 15)
sns.tsplot(data=q1mg0ts_df, time="update", unit="rep",
           condition="treatment", value="score", color=q1_mg0_cmap)
```

### 16 env (NAND, NOT, ORN, AND) randomly changing, Min Genome 100

```python
# # Plot score over time
fig, ax = plt.subplots()
fig.set_size_inches(30, 15)
sns.tsplot(data=q2mg100ts_df, time="update", unit="rep",
           condition="treatment", value="score", color=q2_mg100_cmap)
```

### 16 env (NAND, NOT, ORN, AND) randomly changing, Min Genome 0

```python
# # Plot score over time
fig, ax = plt.subplots()
fig.set_size_inches(30, 15)
sns.tsplot(data=q2mg0ts_df, time="update", unit="rep",
           condition="treatment", value="score", color=q2_mg0_cmap)
```
